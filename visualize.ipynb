```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('features.csv')
```

```python
sns.scatterplot(data=data, x = data.index, y= data['mean_f0'], hue='label')
plt.xlabel('gender')
plt.grid(axis='y', alpha=0.3)
plt.title('So sánh mean_f0 giữa nam và nữ.')
```
